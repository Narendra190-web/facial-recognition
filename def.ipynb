```python
import cv2
import os
import face_recognition
import pickle

student_dir = r"C:\Users\naren\Desktop\python project\imgs"
image_files = os.listdir(student_dir)
images = []
ids = []

for fn in image_files:
    img = cv2.imread(os.path.join(student_dir, fn))
    if img is None:
        print(f"Unable to load image {fn}")
        continue
    images.append(img)
    ids.append(os.path.splitext(fn)[0])

def findEncodings(image_list):
    encode_list = []
    for img in image_list:
        color = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encodings = face_recognition.face_encodings(color)
        if encodings:
            encode_list.append(encodings[0])
        else:
            print("No face found in one image, skipping.")
    return encode_list

print("Encoding faces......!!")
encodelistknown = findEncodings(images)

if len(encodelistknown) != len(ids):
    print("Number of encodings and IDs mismatch!")

save = [encodelistknown, ids]

with open("encoded file.p", 'wb') as f:
    pickle.dump(save, f)

print(encodelistknown)

print("Encoding saved successfully.")
```
